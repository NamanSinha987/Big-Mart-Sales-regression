```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
train_df=pd.read_csv('Train.csv')
```

```python
test_df=pd.read_csv('Test.csv')
```

```python
train_df.head()
```

```python
train_df.info()
```

```python
test_df.info()
```

```python
#Processing both training and test data into 1 dataset could minimize work
train_df['source']='train'
test_df['source']='test'
data=pd.concat([train_df,test_df],ignore_index=True)
```

```python
data.head(20)
```

```python
data.info()
```

```python
data.describe()
```

```python
data.apply(lambda x: len(x.unique()))
```

```python
#Imputing missing item weight by mean weight of each identifier
data['Item_Weight']=data.groupby("Item_Identifier").transform(lambda x:x.fillna(x.mean()))
data.info()
```

```python
data['Item_Weight'].head(20)
```

```python
data["Outlet_Size"].value_counts()
```

```python
data["Outlet_Size"].fillna("Medium",inplace=True)
```

```python
data["Outlet_Size"].value_counts()
```

```python
data.head(5)
```

```python
# Establishment year should be expressed in form that how old is the outlet
data["Outlet_Establishment_Year"]=2013-data["Outlet_Establishment_Year"]
data.head(5)
```

```python
#encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
#New variable for outlet
data['Outlet'] = le.fit_transform(data['Outlet_Identifier'])
var_mod = ['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type','Outlet_Type','Outlet']
le = LabelEncoder()
for i in var_mod:
    data[i] = le.fit_transform(data[i])
```

```python
data.head()
```

```python
#onehotencoding
data=pd.get_dummies(data, columns=['Item_Fat_Content','Outlet_Location_Type','Outlet_Size','Item_Type','Outlet_Type','Outlet'])
data.head(5)
```

```python
data.dtypes
```

```python
#Dividing back into training and test sets

train = data.loc[data['source']=="train"]
test = data.loc[data['source']=="test"]

test.drop(['Item_Outlet_Sales','source'],axis=1,inplace=True)
train.drop(['source'],axis=1,inplace=True)
train.head()
```

```python
train_y=train["Item_Outlet_Sales"]
train_x=train.drop(['Item_Outlet_Sales','Item_Identifier','Outlet_Identifier'],axis=1)
train_x=train_x.values

train_x
```

```python
test_x=test.drop(['Item_Identifier','Outlet_Identifier'],axis=1).values
test_x
```

```python
#creating linear regression model 
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(train_x,train_y)
```

```python
ypred_reg=regressor.predict(test_x)
ypred_reg
```

```python
#creating submission file
result=pd.DataFrame()
result['Item_Identifier'] = test['Item_Identifier']
result['Outlet_Identifier'] = test['Outlet_Identifier']

result["Item_Outlet_Sales"] = ypred_reg
result = result.sort_index()
result.to_csv('submission.csv',index=False)
```
